```python
import datetime

datetime_object = datetime.datetime.now()

class CellPhone():
    '''This entity defines the Cellphone's blueprint'''
    
    def __init__(self, color, model, storage, carrier, phone):
        '''This is a parametrized constructor which assigns all the parametric values to hidden variables'''
        self.__color   = color
        self.__model   = model
        self.__storage = storage
        self.__carrier = carrier
        self.__phone   = phone
        print("This is a {} {} {} Phone\nCapacity is {} GB. Carrier is {}".format(self.__color, self.__model, self.__phone, self.__storage, self.__carrier))
    
    '''
    Accessor Methods 
    
    '''
    
    def get_color(self):
        return self.__color
    
    def get_model(self):
        return self.__model
    
    def get_storage(self):
        print("Storage is {} GB".format(self.__storage))
        
    def get_carrier(self):
        return self.__carrier
              
    '''
    Mutator Methods 
    
    '''   
        
    def set_color(self, color):
        self.__color = color
       
    def set_model(self, model):
        self.__model = model
    
    def set_storage(self, storage):
        self.__storage = storage
          
    def set_carrier(self, carrier):
        self.__carrier = carrier
                
        
    
    def get_time(self):
        '''This method returns the current time'''
        print("Time is: "+datetime_object.strftime("%I:%M %p"))
    
    def get_date(self):
        '''This method returns the current date'''
        print("Date is: "+datetime_object.strftime("%d/%m/%Y"))
    
    def place_call(self, number):
        '''This method prints the parametric phone number'''
        print("Calling number {}...".format(number))
    
    def send_sms(self, msg, number):
        '''This method prints the message as wall as the number'''
        print("Sending Message \'{}\' to number {}".format(msg,number))
        
class ApplePhone(CellPhone):
    '''This entity defines the ApplePhone's blueprint'''
    
    def __init__(self, color, model, storage, carrier):
        '''This is a parametrized constructor which calls the super class's cosntructor'''
        super().__init__(color, model, storage, carrier,'Apple')  
            
    '''
    Accessor Methods 
    
    '''   
    def get_airdrop_status(self):
        return self.__airdrop_status
            
    '''
    Mutator Methods 
    
    '''   
    def set_airdrop_status(self, airdrop_status):
        self.__airdrop_status = airdrop_status
    
    
    
    def launch_Siri(self):
        '''This method prints the Launching siri'''
        print("Launching Siri...")
        
    def upgrade_storage(self,storage):
        '''This method prints the error message in case of Apple Phones storage upgraation'''
        print("Updating storage is not allowed on Apple phones")
        
    def connect_cloud(self):
        '''This method prints the Connecting to iCloud...'''
        print("Connecting to iCloud...")
        
    def download_app(self, app):
        '''This method prints the parametrized app'''
        print("Opening Apple Store to download {} app...".format(app))
    
    def __str__(self):
        '''This method prints the below statement whenever the object of this class is printed'''
        return "AirDrop Status is {}".format(self.__airdrop_status)
    
class AndroidPhone(CellPhone):
    '''This entity defines the ApplePhone's blueprint'''
    
    def __init__(self, color, model, storage, carrier):
        '''This is a parametrized constructor which calls the super class's cosntructor'''
        super().__init__(color, model, storage, carrier,'Android')
             
    '''
    Accessor Methods 
    
    '''      
    def get_split_screen_mode(self):
        return self.__split_screen_mode
            
    '''
    Mutator Methods 
    
    '''   
    def set_split_screen_mode(self, split_screen_mode):
        self.__split_screen_mode = split_screen_mode
        
    def launch_google_assistant(self):
        '''This method prints the Launching Google Assistant...'''
        print("Launching Google Assistant...")
        
    def upgrade_storage(self,storage):
        '''This method prints the new memory value in case of upgradation else it displays error saying You can't rollback memory!'''
        if storage > self._CellPhone__storage:
            print("Old storage is {} GB\nNew storage is {} GB".format(self._CellPhone__storage,storage))
            super().set_storage(storage)
        else:
            print("You can't rollback memory!")
    
    def connect_cloud(self):
        '''This method prints the Connecting to gdrive...'''
        print("Connecting to gdrive...")
        
    def download_app(self, app):
        '''This method prints the parametrized app'''
        print("Opening Google Play to download {} app...".format(app))
    
    def __str__(self):
        '''This method prints the below statement whenever the object of this class is printed'''
        return "Split-Screen Mode is {}".format(self.__split_screen_mode)
    
iPhone = ApplePhone('Black', '11 Pro', 64, 'Verizon') 
iPhone.set_airdrop_status('On');
print(iPhone)
iPhone.get_storage()
iPhone.upgrade_storage(128)
iPhone.connect_cloud()
iPhone.download_app('Starbucks')
iPhone.get_time()
iPhone.get_date()
iPhone.place_call('425-123-4567')
iPhone.send_sms('Hi There', '214-214-2144')
iPhone.launch_Siri()
print('AirDrop Status is: ', iPhone.get_airdrop_status())

print()
print()

aPhone = AndroidPhone('White', 'Note 10', 128, 'AT&T') 
aPhone.set_split_screen_mode('On');
print(aPhone)
aPhone.get_storage()
aPhone.upgrade_storage(256)
aPhone.upgrade_storage(128)
aPhone.get_storage()
aPhone.connect_cloud()
aPhone.download_app('WhatsApp')
aPhone.get_time()
aPhone.get_date()
aPhone.place_call('206-123-4567')
aPhone.send_sms('Hi There', '469-123-4567')
aPhone.launch_google_assistant()
print('Split-Screen Mode is: ', aPhone.get_split_screen_mode())
```

```python
class Operations:
    '''This class handles calculator operations'''
    
    def addition(self, number_1, number_2):
        '''This method is for calculating addition of two numbers
        
        Parameters: It accepts two parameters.
        Return Value: It returns addition of two numbers.
        
        '''
        return number_1+number_2
    
    def subtraction(self, number_1, number_2):
        '''This method is for calculating subtraction of two numbers
                
        Parameters: It accepts two parameters.
        Return Value: It returns subtraction of two numbers.
        
        '''
        return number_1-number_2
    
    def multiplication(self, number_1 ,number_2):
        '''This method is for calculating multiplication of two numbers        
        Parameters: It accepts two parameters.
        Return Value: It returns multiplication of two numbers.
        
        '''
        return number_1*number_2
    
    def division(self, number_1, number_2):
        '''This method is for calculating division of two numbers        
        Parameters: It accepts two parameters.
        Return Value: It returns division of two numbers.
        
        '''
        return number_1/number_2

class Display:
    '''This class handles diplay of all the calculated outputs'''
    
    def __init__(self, number_1, number_2):
        self.__number_1 = number_1
        self.__number_2 = number_2
        self.__op       = Operations()
    
    def display_addition(self):  
        '''This method is for displaying addition of two numbers'''
        print(self.__number_1, "+", self.__number_2, "=", self.__op.addition(self.__number_1, self.__number_2))
            
    def display_subtraction(self):
        '''This method is for displaying subtraction of two numbers'''
        print(self.__number_1, "-", self.__number_2, "=", self.__op.subtraction(self.__number_1, self.__number_2))
        
    def display_multiplication(self):
        '''This method is for displaying multiplication of two numbers'''
        print(self.__number_1, "*", self.__number_2, "=", self.__op.multiplication(self.__number_1, self.__number_2))
        
    def display_division(self):
        '''This method is for displaying division of two numbers'''
        print(self.__number_1, "/", self.__number_2, "=", self.__op.division(self.__number_1, self.__number_2))
        
        
class Calculator:
    '''This is the main Calculator class which calls all the other classes and their operations based on the input''' 
    
    def __init__(self, number_1, number_2, opertn):
        '''This is a parametrized constructor which assigns three values number_1, number_2, and opertn and also calls the Display class's parametrized constructor.'''
        self.__number_1 = number_1
        self.__number_2 = number_2
        self.__opertn   = opertn
        self.__dis      = Display(self.__number_1, self.__number_2)
            
    def calculate(self):
        '''This fuctions checks the input opeartion parameter and calls appropriate functions of Display class'''
        if self.__opertn   == '+':
            self.__dis.display_addition()

        elif self.__opertn == '-':
            self.__dis.display_subtraction()

        elif self.__opertn == '*':
            self.__dis.display_multiplication()

        elif self.__opertn == '/':
            self.__dis.display_division()
        else:
            print("Invalid input")

a = Calculator(2,3,'+')
a.calculate()

b = Calculator(10,3,'-')
b.calculate()

c = Calculator(4,3,'*')
c.calculate()

d = Calculator(10,5,'/')
d.calculate()
```
